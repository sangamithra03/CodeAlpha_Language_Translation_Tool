```python
import gradio as gr
from deep_translator import GoogleTranslator
```

```python
def translate_text(text, source_lang, target_lang):
    try:
        translated = GoogleTranslator(
            source=source_lang,
            target=target_lang
        ).translate(text)
        return translated
    except Exception as e:
        return f"Error: {e}"
```

```python
interface = gr.Interface(
    fn=translate_text,
    inputs=[
        gr.Textbox(
            lines=4,
            placeholder="Enter text here...",
            label="Enter Text"
        ),
        gr.Dropdown(
            choices=["auto", "en", "ta", "hi", "fr", "es", "de"],
            value="auto",
            label="Source Language"
        ),
        gr.Dropdown(
            choices=["en", "ta", "hi", "fr", "es", "de"],
            value="ta",
            label="Target Language"
        )
    ],
    outputs=gr.Textbox(label="Translated Text"),
    title="🌍 AI Language Translation Tool",
    description="Translate text between multiple languages using Google Translator."
)

interface.launch()
```
